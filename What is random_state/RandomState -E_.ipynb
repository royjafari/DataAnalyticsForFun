# Random_state?

```python
import pandas as pd
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
```

```python
customer_df = pd.read_csv('Customer Churn.csv')
customer_df
```

```python
select_attributes = ['Call  Failure', 'Complains', 'Charge  Amount',
                     'Frequency of SMS', 'Distinct Called Numbers', 'Age',
                     'Customer Value']
customer_df = pd.DataFrame(customer_df[select_attributes])
```

```python
customer_df.columns 
```

# Problem
Form clusters of customers that maximizes silhouette_score. We want the customers to be grouped into 14 clusters.

```python
customer_df_nor = (customer_df - customer_df.min())/(customer_df.max() - customer_df.min())
```

```python
kmeans = KMeans(n_clusters=14)
kmeans.fit(customer_df_nor)

silhouette_score(customer_df_nor,kmeans.labels_)
```

```python
for i in range(1,20):
    kmeans = KMeans(n_clusters=14)
    kmeans.fit(customer_df_nor)

    print(silhouette_score(customer_df_nor,kmeans.labels_))
```

```python
for i in range(1,20):
    kmeans = KMeans(n_clusters=14,random_state=i)
    kmeans.fit(customer_df_nor)

    print(i,silhouette_score(customer_df_nor,kmeans.labels_))
```

```python
kmeans = KMeans(n_clusters=14,random_state=17)
kmeans.fit(customer_df_nor)

silhouette_score(customer_df_nor,kmeans.labels_)
```

```python
customer_df['Cluster'] = kmeans.labels_
customer_df
```
